## Data preprocessing

```python
# high_prices = df["High"]
high_prices = df.loc[:,'High'].values
low_prices = df.loc[:, "Low"].values
mid_prices = (high_prices + low_prices) / 2
mid_prices
prices = df[["Open", "High", 'Low', "Close"]]
prices.to_numpy().shape
```

```python
#normalize the data
debug = True

scaler = MinMaxScaler()
train_data = mid_prices[:7000].reshape(-1,1)
test_data = mid_prices[7000:].reshape(-1,1)
if debug:
    print(train_data)
    print(test_data)
smoothing_window_size = 2300
for di in range(0,7000,smoothing_window_size):
    data_slice = train_data[di:di+smoothing_window_size,:]
    if debug:
        print("shape of train_data", data_slice.shape)
#         print("nromalized: ", train_data)
    scaler.fit(data_slice)
    train_data[di:di+smoothing_window_size,:] = scaler.transform(train_data[di:di+smoothing_window_size,:])
# You normalize the last bit of remaining data
# scaler.fit(train_data[di+smoothing_window_size:,:])
# train_data[di+smoothing_window_size:,:] = scaler.transform(train_data[di+smoothing_window_size:,:])
# Reshape both train and test data
train_data = train_data.reshape(-1)

# Normalize test data
print(test_data)
test_data = scaler.transform(test_data).reshape(-1)
```

```python

scaler = MinMaxScaler()
test_data = mid_prices[7000:].reshape(-1,1)
smoothing_window_size = 2300
for di in range(0,1800,1000):
    data_slice = test_data[di:di+smoothing_window_size,:]
    if debug:
        print("shape of test_data", data_slice.shape)
#         print("nromalized: ", train_data)
    scaler.fit(data_slice)
    test_data[di:di+smoothing_window_size,:] = scaler.transform(test_data[di:di+smoothing_window_size,:])
test_data = test_data.reshape(-1)
```

```python
# Now perform exponential moving average smoothing
# So the data will have a smoother curve than the original ragged data
EMA = 0.0
gamma = 0.1
for ti in range(7000):
    EMA = gamma*train_data[ti] + (1-gamma)*EMA
    train_data[ti] = EMA

# Used for visualization and test purposes
all_mid_data = np.concatenate([train_data, test_data],axis=0)
```

```python
# def build_timeseries(mat, y_col_index):
#     # y_col_index is the index of column that would act as output column
#     # total number of time-series samples would be len(mat) - TIME_STEPS
#     # N is number of 
#     # D is number of features
#     # T number of time ste
#     N = mat.shape[0] - TIME_STEPS
#     D = mat.shape[1]

def build_timeseries(mat, y_col_index, TIME_STEPS):
    # y_col_index is the index of column that would act as output column
    # total number of time-series samples would be len(mat) - TIME_STEPS
    dim_0 = mat.shape[0] - TIME_STEPS
    dim_1 = mat.shape[1]
    x = np.zeros((dim_0, TIME_STEPS, dim_1))
    y = np.zeros((dim_0,))
    
    for i in range(dim_0):
        x[i] = mat[i:TIME_STEPS+i]
        y[i] = mat[TIME_STEPS+i, y_col_index]
    print("length of time-series i/o",x.shape,y.shape)
    return x, y

def trim_dataset(mat, batch_size):
    """
    trims dataset to a size that's divisible by BATCH_SIZE
    """
    no_of_rows_drop = mat.shape[0]%batch_size
    print("number of rows dropped", no_of_rows_drop)
    if(no_of_rows_drop > 0):
        return mat[:-no_of_rows_drop]
    else:
        return mat
```

```python
BATCH_SIZE = 100
TIME_STEPS = 100
lr = 0.001
model = Sequential()
model.add(LSTM(100, batch_input_shape=(BATCH_SIZE, TIME_STEPS, train_data.shape[1]), dropout=0.0, recurrent_dropout=0.0, stateful=True, kernel_initializer='random_uniform', return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(60, dropout=0.0))
model.add(Dropout(0.4))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
optimizer = optimizers.RMSprop(lr=lr)
model.compile(loss='mean_squared_error', optimizer=optimizer)
plot_model(model, to_file='model.png')
```

```python
def make_prediction(model, test_data, TIME_STEPS, BATCH_SIZE, df):
    """
    the test_data passed in is the matrix of features with shape 
    (T, D) where T is the number of time steps there are and D is 
    the number of features
    """
    scale = MinMaxScaler()
    temp = scale.fit_transform(test_data)
    print("shape1:", temp.shape)
    temp, _ = build_timeseries(temp, 3, TIME_STEPS)
#     temp = trim_dataset(temp, BATCH_SIZE)
    print("shape:", temp.shape)
    prediction = model.predict(temp)
    pred = (prediction* scale.data_range_[0]) + scale.data_min_[0]
    print("prediction shape:", pred.shape)
    figure(num=None, figsize=(15, 6), dpi=70, facecolor='w', edgecolor='k')
    plt.plot(df.reset_index()['Date'].values[:len(pred)],pred)
    plt.plot(df.reset_index()['Date'].values[:len(pred)],df['Close'].values[TIME_STEPS:TIME_STEPS +  len(pred)])

    plt.tight_layout()
    plt.grid()
    plt.show()
    return pred
```
